```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = load_iris()
```

```python
X = data["data"]
y = data["target"]
```

```python
X = X/np.sum(np.abs(X), axis=0)
```

```python
df = pd.DataFrame(X)
df["Target"] = y
```

```python
sns.pairplot(df, hue="Target")
```

```python
plt.scatter(X[y==0, 2], X[y==0, 3])
plt.scatter(X[y==1, 2], X[y==1, 3])
plt.scatter(X[y==2, 2], X[y==2, 3])
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X[:, 2:4], labels=y)
```

```python
D = np.linalg.norm(X[np.newaxis, :, :] - X[:, np.newaxis, :], axis=2)
```

```python
silhouette_score(D, labels=y, metric="precomputed")
```
